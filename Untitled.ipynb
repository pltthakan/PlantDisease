```python
# === 1. KÜTÜPHANELER ===
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
import zipfile, json, os, random
from google.colab import files
from collections import Counter

# === 2. SABİTLER ve RANDOM SEED ===
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS_1 = 5
EPOCHS_2 = 25
LR_1 = 1e-3
LR_2 = 1e-5
base_dir = 'dataset/PlantVillage'

# === 3. KONTROLLER ===
print("TensorFlow Sürümü:", tf.__version__)
if tf.config.list_physical_devices('GPU'):
    print("--- GPU Aktif --- (Harika!)")
else:
    print("--- DİKKAT: GPU Aktif Değil! ---")
    print("Lütfen 'Çalışma Zamanı > Çalışma zamanı türünü değiştir > GPU' seçin ve bu hücreyi yeniden çalıştırın.")

# === 4. KAGGLE VERİSİ ===
!pip install -q kaggle
uploaded = files.upload()
if 'kaggle.json' in uploaded:
    !mkdir -p ~/.kaggle
    !cp kaggle.json ~/.kaggle/
    !chmod 600 ~/.kaggle/kaggle.json
    !kaggle datasets download -d emmarex/plantdisease -p dataset
    with zipfile.ZipFile('dataset/plantdisease.zip', 'r') as zip_ref:
        zip_ref.extractall('dataset')
else:
    print("Kaggle kimlik dosyası yüklenmedi.")



# === 5. VERİ YÜKLEYİCİLER ===
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=30,
    width_shift_range=0.2, height_shift_range=0.2,
    zoom_range=0.2, horizontal_flip=True,
    validation_split=0.2
)
val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2
)

train_gen = train_datagen.flow_from_directory(
    base_dir, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='training',
    shuffle=True, seed=SEED
)
val_gen = val_datagen.flow_from_directory(
    base_dir, target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='validation',
    shuffle=False, seed=SEED
)

NUM_CLASSES = len(train_gen.class_indices)
with open('class_indices.json', 'w') as f:
    json.dump(train_gen.class_indices, f)
print(f"\nToplam {NUM_CLASSES} sınıf bulundu.")

# Sınıf adlarını model/rapor için sabitle
class_names = [k for k,_ in sorted(train_gen.class_indices.items(), key=lambda x: x[1])]
print("Sınıf sırası:", class_names)

# === 6. MODEL OLUŞTURMA ===
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_model.trainable = False

x = GlobalAveragePooling2D()(base_model.output)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)
out = Dense(NUM_CLASSES, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=out)

model.compile(optimizer=Adam(LR_1), loss='categorical_crossentropy', metrics=['accuracy'])

# === 7. CALLBACK'LER ===
callbacks = [
    ModelCheckpoint('mobilenet_stage1.keras', monitor='val_accuracy', save_best_only=True, mode='max'),
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, verbose=1)
]

# === 8. AŞAMA 1: Kafa Eğitimi ===
print("\n--- AŞAMA 1: Kafa Eğitimi ---")
history1 = model.fit(train_gen, validation_data=val_gen, epochs=EPOCHS_1, callbacks=callbacks)

# === 9. AŞAMA 2: Fine-Tuning ===
print("\n--- AŞAMA 2: İnce Ayar ---")
base_model.trainable = True
fine_tune_at = len(base_model.layers) - 30  # son ~30 katmanı aç
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(optimizer=Adam(LR_2), loss='categorical_crossentropy', metrics=['accuracy'])

callbacks_ft = [
    ModelCheckpoint('mobilenet_finetuned.keras', monitor='val_accuracy', save_best_only=True, mode='max'),
    EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, verbose=1)
]

# (Opsiyonel) Sınıf dengesizliği için class_weight
counts = Counter(train_gen.classes)
max_count = max(counts.values())
class_weight = {c: max_count/n for c, n in counts.items()}
print("class_weight:", class_weight)

history2 = model.fit(
    train_gen, validation_data=val_gen,
    epochs=EPOCHS_2, callbacks=callbacks_ft,
    class_weight=class_weight  # dengesizlik yoksa istersen kaldır
)

# === 10. GRAFİKLER ===
acc = history1.history.get('accuracy', []) + history2.history.get('accuracy', [])
val_acc = history1.history.get('val_accuracy', []) + history2.history.get('val_accuracy', [])
plt.figure(figsize=(10,5))
plt.plot(acc, label='Eğitim')
plt.plot(val_acc, label='Doğrulama')
plt.title("MobileNetV2 Eğitim Süreci")
plt.xlabel("Epochs"); plt.ylabel("Doğruluk"); plt.legend(); plt.show()

# === 11. EN İYİ MODELİN YÜKLENMESİ ===
model = tf.keras.models.load_model('mobilenet_finetuned.keras')
loss, acc = model.evaluate(val_gen, verbose=0)
print(f"\nSon doğrulama doğruluğu: %{acc*100:.2f}")

model.save('bitki_hastalik_mobilenet_final.keras')
print("✅ Nihai model kaydedildi: bitki_hastalik_mobilenet_final.keras")

# === 12. DEĞERLENDİRME RAPORU ===
val_gen.reset()
preds = model.predict(val_gen, verbose=0)
y_pred = np.argmax(preds, axis=1)

print("\n--- SINIFLANDIRMA RAPORU ---")
print(classification_report(val_gen.classes, y_pred, target_names=class_names))

cm = confusion_matrix(val_gen.classes, y_pred)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=False, cmap="Greens",
            xticklabels=class_names, yticklabels=class_names)
plt.ylabel('Gerçek'); plt.xlabel('Tahmin'); plt.title('Confusion Matrix')
plt.show()

# === 13. TEK GÖRSEL TAHMİNİ ===
img_path = 'test_leaf.jpeg'  # test fotoğrafı
if os.path.exists(img_path):
    img = load_img(img_path, target_size=(224, 224))
    arr = img_to_array(img)
    arr = np.expand_dims(arr, axis=0)
    arr = preprocess_input(arr)  # Eğitimle aynı ölçek
    pred = model.predict(arr, verbose=0)
    predicted_class = int(np.argmax(pred))
    with open('class_indices.json', 'r') as f:
        labels = json.load(f)
    inv_labels = {v:k for k,v in labels.items()}
    print(f"Tahmin edilen sınıf: {inv_labels[predicted_class]} (%{pred[0][predicted_class]*100:.2f})")

# === 14. GRAD-CAM ===
def make_gradcam_heatmap(img_array, model, last_conv_layer_name):
    grad_model = tf.keras.models.Model([model.inputs],
                                       [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        pred_index = tf.argmax(predictions[0])
        class_channel = predictions[:, pred_index]
    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs), axis=-1)
    heatmap = np.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

if os.path.exists(img_path):

    heatmap = make_gradcam_heatmap(arr, model, 'Conv_1')  # MobileNetV2'nin son conv katmanı
    plt.matshow(heatmap)
    plt.title("Grad-CAM ısı haritası")
    plt.show()
```
